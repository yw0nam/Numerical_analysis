# Open Method

## Declare function for SFP method  Y = exp(-2x) / 7

```python
Y_SFP = lambda x:  np.exp(-2 * x) / 7
```

## Simple Fixed Point(SFP) method

```python
def simple_fixed_point(f, fin_val, x_init):
    i = 1  #iteration 
    temp = 0 # Variable for store previous root, and first temp value should be 0
    error_percentage = 100    #First error percentage should be 100
    while(error_percentage > fin_val):
        
        if(i == 1):
            next_x = f(x_init)                                  # For First trial
        elif(i != 1):
            temp = next_x                                       # Store previous X
            next_x = f(next_x)                                  # Determine next X
        error_percentage = abs((next_x - temp) / next_x) * 100  # Calculate error rate
    
        print("trial:{0} \t X:{1:.4e} \t X_before:{2:.4e} \t error:{3:.4e}\n"
              .format(i, next_x, temp, error_percentage), '-' * 80)
        if (i == 1000):
            print("Can't find root")
            break
        i = i + 1
```

```python
simple_fixed_point(Y_SFP, 10 ** -6, 3)
```

```python
simple_fixed_point(Y_SFP, 10 ** -6, -30)
```

## Declare function for Newton-Raphson 
## Y = exp(-2X) - 7X   ,      Y' = -2exp(-2X) - 7

```python
Y_NR = lambda x: np.exp(-2 * x ) - 7 * x
Y_NR_derivation = lambda x: -2 * np.exp(-2 * x) - 7
```

## Newton Raphson Method

```python
def newton_Raphson(f, f_der, fin_val, x_init):
    i = 1  #iteration 
    temp = 0 # Variable for store previous root, and first temp value should be 0
    error_percentage = 100       #First error percentage should be 100
    while(error_percentage > fin_val):
        
        if(i == 1):
            next_x = x_init - f(x_init) / f_der(x_init)         # For First trial
        elif(i != 1):
            temp = next_x                                       # Store previous X
            next_x =  next_x - f(next_x) / f_der(next_x)        # Determine next X
        error_percentage = abs((next_x - temp) / next_x) * 100  # Calculate error rate
    
        print("trial:{0}\t X:{1:.4e}\t X_before:{2:.4e}\t error:{3:.4e}\n"
              .format(i, next_x, temp, error_percentage), '-' * 80)
        if (i == 1000):
            print("Can't find root")
            break
        i = i + 1
```

```python
newton_Raphson(Y_NR, Y_NR_derivation, 10 ** -6, 3)
```

```python
newton_Raphson(Y_NR, Y_NR_derivation, 10 ** -6, -30)
```

## Declare function for Modifed Secant Y = exp(-2X) - 7X

```python
del_x = 10 ** -6
Y_M_S = lambda x: np.exp(-2 * x) - 7 * x
```

## Modifed Secant method

```python
def modfied_secant(f, delta_x, fin_val, x_init):
    i = 1  #iteration 
    temp = 0 # Variable for store previous root, and first temp value should be 0
    error_percentage = 100       #First error percentage should be 100
    while(error_percentage > fin_val):
        
        if(i == 1):
            next_x = x_init - delta_x * f(x_init) / (f(x_init + delta_x) - f(x_init))          # For First trial
        elif(i != 1):
            temp = next_x                                                                      # Store previous X
            next_x =  next_x - (delta_x * f(next_x)) / (f(next_x + delta_x) - f(next_x))                                       # Determine next X
        error_percentage = abs((next_x - temp) / next_x) * 100                                 # Calculate error rate
    
        print("trial:{0}\t X:{1:.4e}\t X_before:{2:.4e}\t error:{3:.4e}\n"
              .format(i, next_x, temp, error_percentage), '-' * 80)
        if (i == 1000):
            print("Can't find root")
            break
        i = i + 1
```

```python
modfied_secant(Y_M_S, del_x, 10 ** -6,  3)
```

```python
modfied_secant(Y_M_S, del_x, 10 ** -6,  -30)
```
